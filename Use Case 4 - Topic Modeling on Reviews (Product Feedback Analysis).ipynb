## Use Case 4 â€” Topic Modeling on Reviews (Product Feedback Analysis)

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
import numpy as np

df = pd.read_excel("C:\Semester 3\Text, Web & Sen Analytics\Capston_Project-Priya_Saini(24MBMB34)\ecom_analytics_data.xlsx")
texts = df['review_text'].fillna("").astype(str)

# Vectorize
vectorizer = TfidfVectorizer(max_df=0.9, min_df=2, stop_words='english')
X = vectorizer.fit_transform(texts)

# NMF topic model
n_topics = 4
nmf = NMF(n_components=n_topics, random_state=42)
W = nmf.fit_transform(X)
H = nmf.components_

feature_names = vectorizer.get_feature_names_out()
def top_words(topic_idx, n=8):
    topn_ids = H[topic_idx].argsort()[:-n-1:-1]
    return [feature_names[i] for i in topn_ids]

for t in range(n_topics):
    print(f"Topic {t}: ", top_words(t))

# Assign topic to each review (dominant component)
topic_assignments = W.argmax(axis=1)
df['topic'] = topic_assignments
print("\nSample reviews by topic:")
for t in range(n_topics):
    print(f"\nTopic {t} example reviews:\n", df[df['topic']==t]['review_text'].head(3).tolist())
```
