```python
import pandas as pd

df = pd.read_excel("C:\Semester 3\Text, Web & Sen Analytics\Capston_Project-Priya_Saini(24MBMB34)\ecom_analytics_data.xlsx")

# Key metrics
metrics = df.groupby(['region','product_category']).agg(
    orders=('order_id','count'),
    avg_delivery_days=('delivery_time_days','mean'),
    on_time_rate=('delivery_status', lambda x: (x=="On Time").mean())
).reset_index()

# Sort by worst on_time_rate
worst_regions = metrics.sort_values(by='on_time_rate').head(10)
print("Worst region-category combos by on-time rate:\n", worst_regions)

# Delivery delay analysis overall
overall = {
    'avg_delivery_days': df['delivery_time_days'].mean(),
    'pct_on_time': (df['delivery_status']=="On Time").mean(),
    'pct_delayed': (df['delivery_status']=="Delayed").mean(),
    'pct_very_delayed': (df['delivery_status']=="Very Delayed").mean()
}
print("\nOverall delivery KPIs:", overall)

# Example drill-down: Are large orders delayed more?
df['big_order'] = df['order_value'] > df['order_value'].median()
print("\nAvg delivery days by big_order:\n", df.groupby('big_order')['delivery_time_days'].mean())
```
