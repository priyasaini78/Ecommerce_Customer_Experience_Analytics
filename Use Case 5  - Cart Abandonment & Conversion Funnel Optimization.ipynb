## Use Case 5 â€” Cart Abandonment & Conversion Funnel Optimization

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

df = pd.read_excel("C:\Semester 3\Text, Web & Sen Analytics\Capston_Project-Priya_Saini(24MBMB34)\ecom_analytics_data.xlsx")

# Abandonment rate by device
print("Abandonment rate by device:\n", df.groupby('device')['cart_abandoned'].mean())

# Abandonment by pages visited / session duration
print("\nAvg pages visited where cart abandoned vs not:")
print(df.groupby('cart_abandoned')['pages_visited'].mean())

# Build a simple model to predict cart_abandoned
df_model = pd.get_dummies(df[['pages_visited','session_duration_sec','promo_used','device','region','cart_abandoned']], drop_first=True)
X = df_model.drop(columns=['cart_abandoned'])
y = df_model['cart_abandoned']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y, test_size=0.25)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
print(classification_report(y_test, pred))

# Feature importance
importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop features for cart abandonment prediction:\n", importances.head(10))
```
